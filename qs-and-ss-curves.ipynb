```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = qsdqsf[:,0]
qsd = qsdqsf[:,1]
qsf = qsdqsf[:,2]
```

```python
plt.plot(t, qsf, label='qsd')
plt.plot(t, qsd, label='qsf')
plt.xlabel('time')
plt.legend(loc='best')
```

```python
plt.title('relative changes of qsf')
plt.plot(t, qsf/qsf[0])
```

```python
plt.title('relative changes of qsd')
plt.plot(t, qsd/qsd[0])
```

```python
def ssd(ke1):
    vssd = -0.04381647074 + ke1*(0.009150934932+ ke1*(-0.0023900895  + ke1*(0.000318068028 -  ke1* 1.6353205e-05)))
    return vssd

def ssf(ke1):
    vssf =  -0.006384940088 + ke1*(0.01967542848 + ke1*( -0.006776746 + ke1*(0.00091367565 - ke1* 4.293705e-05)))
    return vssf
```

```python
ke = 0.4 + (8.0 - 0.4) * np.arange(1000.0)/1000
```

```python
plt.title('ssd')
plt.plot(ke, ssd(ke))
plt.xlabel('kinetic energy [GeV]')
```

```python
plt.title('relative changes of ssd')
plt.plot(ke, ssd(ke)/ssd(ke[0]))
plt.xlabel('kinetic energy [GeV]')
```

```python
plt.title('ssf')
plt.plot(ke, ssf(ke), label='full')
plt.xlabel('kinetic energy')
plt.legend(loc='best')
```

```python
plt.title('ssf')
#plt.plot(ke, ssf(ke), label='full')
plt.plot(ke, -0.006384940088 + ke*0.01967542848, label='linear')
plt.plot(ke, -0.006384940088 + ke*(0.01967542848 + ke*( -0.006776746)), label='quadratic')
plt.plot(ke,  -0.006384940088 + ke*(0.01967542848 + ke*( -0.006776746 + ke*(0.00091367565 ))) , label='cubic')
plt.plot(ke, -0.006384940088 + ke*(0.01967542848 + ke*( -0.006776746 + ke*(0.00091367565 - ke* 4.293705e-05))) , label='quartic')
plt.xlabel('kinetic energy')
plt.legend(loc='best')
```
